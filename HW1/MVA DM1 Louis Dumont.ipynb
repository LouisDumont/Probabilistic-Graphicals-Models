#### Import libraries and data

```python
# Usefull libraries

import numpy as np
import os
import matplotlib.pyplot as plt
from math import log, exp
from scipy.optimize import minimize

# Path to the data
cur_path = os.getcwd()
data_path = cur_path + "/homework1_data/data/"


# Loading the data

def load_data(path):
    mat = np.loadtxt(path, delimiter=' ', converters={2:lambda x: float(chr(x[-2]))})
    features = mat[:,:-1]
    labels = mat[:,-1]
    return features, labels.astype(bool)

dataset = "C"

features, labels = load_data(data_path+"train"+dataset)
test_features, test_labels = load_data(data_path+"test"+dataset)
```

## 1-  Generative Model (LDA)

```python
# Computing the MLE for this model
 

def mle_LDA(features, labels):
    pi = np.sum(labels)/labels.size
    mu0 = np.mean(features[np.logical_not(labels)], axis=0)
    mu1 = np.mean(features[labels], axis=0)
    sigma = np.zeros((2,2))
    for i in range(labels.size):
        if not labels[i]:
            vec = (features[i]-mu0).reshape(-1, 1)
        else:
            vec = (features[i]-mu1).reshape(-1, 1)
        sigma += np.dot(vec,np.transpose(vec))
    sigma = sigma / labels.size
    
    return pi, mu0, mu1, sigma
```

```python
# Plot the decision boundary linked to it

# Compute relevantmatrices and vectors

pi, mu0, mu1, sigma = mle_LDA(features, labels)

sigma_inv = np.linalg.inv(sigma)

beta = np.dot(sigma_inv, (mu1-mu0)) 
gamma = -0.5 * np.dot(np.transpose(mu1+mu0), beta) - log((1-pi)/pi) 

# Compute the decision boundary
# the decision function is given by beta * x = -gamma

def set_x1(x0):
    return (-gamma - beta[0]*x0)/beta[1]

set_x1_vect = np.vectorize(set_x1)

x0_values = np.linspace(np.amin(features[:,0]), np.amax(features[:,0]), 5000)
x1_values = set_x1_vect(x0_values)

# Plot the decision boundary

fig = plt.figure()

plt.scatter(features[labels,0], features[labels,1], c='blue')
plt.scatter(features[np.logical_not(labels),0], features[np.logical_not(labels),1], c='red')
plt.scatter(x0_values, x1_values, s=1, c='green')

plt.show()
```

## 2- Logistic Regression

```python
# Utils functions
def logistic(z):
    return 1/(1+exp(-z))

def linear(w,x,b):
    return np.dot(np.transpose(w), x) + b

# Compute the log-likelihood of the sample for a given set of parameters
def log_likelihood(features, labels, w, b):
    res = 0
    for i in range(labels.size):
        x = features[i]
        if labels[i]:
            res += log(logistic(linear(w,x,b)))
        else:
            res +=log(1 - logistic(linear(w,x,b)))
    return res

# Functions to minimize to get the MLE
def func_to_min(params, features, labels):
    return -log_likelihood(features, labels, np.array(params[:-1]), params[-1])

# Computes the w and b corresponding to the MLE
def learn_params_LogReg(features, labels):
    mle = minimize(lambda x:func_to_min(x, features, labels), [np.ones(features[0].shape[0]+1)])
    return mle.x[:-1], mle.x[-1]
 
w,b = learn_params_LogReg(features, labels)

#mle = minimize(func_to_min, [np.ones(features[0].shape[0]+1)])
#print(res.x)
print("Learned parameters for the logistic regression:")
print("w:", w)
print("b:", b)
```

```python
# Compute the decision function boundary

def set_x1(x0):
    return (-b - w[0]*x0)/w[1]

set_x1_vect = np.vectorize(set_x1)

x0_values = np.linspace(np.amin(features[:,0]), np.amax(features[:,0]), 5000)
x1_values = set_x1_vect(x0_values)

# Plot the decision function boundary

fig = plt.figure()

plt.scatter(features[labels,0], features[labels,1], c='blue')
plt.scatter(features[np.logical_not(labels),0], features[np.logical_not(labels),1], c='red')
plt.scatter(x0_values, x1_values, s=1, c='green')

plt.show()
```

## 3- Linear Regression

```python
# Learn the parameters

def learn_params_LinReg(features, labels):
    X = np.stack((features[:,0],features[:,1] , np.ones(features.shape[0])), axis=1)
    
    term1 = np.linalg.inv(np.dot(np.transpose(X), X))
    term2 = np.dot(np.transpose(X), labels)

    w_tilde = np.dot(term1, term2)

    w = w_tilde[:-1]
    b = w_tilde[-1]
    
    return w,b

w, b = learn_params_LinReg(features, labels)

print("Learned parameters for the logistic regression:")
print("w:", w)
print("b:", b)

# Commpute the decision boundary

def set_x1(x0):
    # Here the decision function corresponds to beta * x + gamma = 0.5
    return (0.5-b - w[0]*x0)/w[1]

set_x1_vect = np.vectorize(set_x1)

x0_values = np.linspace(np.amin(features[:,0]), np.amax(features[:,0]), 5000)
x1_values = set_x1_vect(x0_values)

# Plot the decision boundary

fig = plt.figure()

plt.scatter(features[labels,0], features[labels,1], c='blue')
plt.scatter(features[np.logical_not(labels),0], features[np.logical_not(labels),1], c='red')
plt.scatter(x0_values, x1_values, s=1, c='green')

plt.show()
```

## 4- Application

```python
# Get the parameters learned by LinReg and LogReg for each dataset

for dataset in ["A", "B", "C"]:
    features, labels = load_data(data_path+"train"+dataset)
    print("Parameters learned for dataset " + dataset)
    w,b = learn_params_LinReg(features, labels)
    print("For Linear Regression: ", w, b)
    w,b = learn_params_LogReg(features, labels)
    print("For Logistic Regression: ", w, b)
```

```python
# Create proper classification function for each method
# They compute the classification of test points given a training set

def classif_point(x, beta, gamma):
    return beta[0]*x[0] + beta[1]*x[1] + gamma >= 0

def classif_LDA(train_features, train_labels, test_features):
    pi, mu0, mu1, sigma = mle_LDA(train_features, train_labels)
    sigma_inv = np.linalg.inv(sigma)
    beta = np.dot(sigma_inv, (mu1-mu0)) 
    gamma = -0.5 * np.dot(np.transpose(mu1+mu0), beta) - log((1-pi)/pi)
        
    return np.apply_along_axis(lambda x: classif_point(x,beta,gamma), 1, test_features)#classifieur(test_features)

def classif_LogReg(train_features, train_labels, test_features):
    w, b = learn_params_LogReg(train_features, train_labels)
    return np.apply_along_axis(lambda x: classif_point(x,w,b), 1, test_features)

def classif_LinReg(train_features, train_labels, test_features):
    w, b = learn_params_LinReg(train_features, train_labels)
    return np.apply_along_axis(lambda x: classif_point(x,w,b-0.5), 1, test_features)
```

```python
# Compute the errors for each model/dataset

for dataset in ["A", "B", "C"]:
    
    train_features, train_labels = load_data(data_path+"train"+dataset)
    test_features, test_labels = load_data(data_path+"test"+dataset)
    
    print("Train errors for dataset " + dataset)
    pred_LDA = classif_LDA(train_features, train_labels, train_features)
    pred_LinReg = classif_LinReg(train_features, train_labels, train_features)
    pred_LogReg = classif_LogReg(train_features, train_labels, train_features)
    print("LDA:", np.sum(np.not_equal(pred_LDA, train_labels))/train_labels.size)
    print("LinReg:", np.sum(np.not_equal(pred_LinReg, train_labels))/train_labels.size)
    print("LogReg:", np.sum(np.not_equal(pred_LogReg, train_labels))/train_labels.size)
    
    print("Test errors for dataset " + dataset)
    pred_LDA = classif_LDA(train_features, train_labels, test_features)
    pred_LinReg = classif_LinReg(train_features, train_labels, test_features)
    pred_LogReg = classif_LogReg(train_features, train_labels, test_features)
    print("LDA:", np.sum(np.not_equal(pred_LDA, test_labels))/test_labels.size)
    print("LinReg:", np.sum(np.not_equal(pred_LinReg, test_labels))/test_labels.size)
    print("LogReg:", np.sum(np.not_equal(pred_LogReg, test_labels))/test_labels.size)
```

## 5- QDA

```python
# Compute the MLE

def mle_QDA(features, labels):
    pi = np.sum(labels)/labels.size
    mu0 = np.mean(features[np.logical_not(labels)], axis=0)
    mu1 = np.mean(features[labels], axis=0)
    sigma0 = np.zeros((2,2))
    sigma1 = np.zeros((2,2))
    
    for i in range(labels.size):
        if not labels[i]:
            vec = (features[i]-mu0).reshape(-1, 1)
            sigma0 += np.dot(vec,np.transpose(vec))
        else:
            vec = (features[i]-mu1).reshape(-1, 1)
            sigma1 += np.dot(vec,np.transpose(vec))
            
    sigma0 = sigma0 / np.sum(1-labels)
    sigma1 = sigma1 / np.sum(labels)
    
    return pi, mu0, mu1, sigma0, sigma1

# derive from the MLE the parameters for the decision function

def learn_params_QDA(features, labels):
    pi, mu0, mu1, sigma0, sigma1 = mle_QDA(features, labels)
    sigma0_inv, sigma1_inv = np.linalg.inv(sigma0), np.linalg.inv(sigma1)
    
    A = (sigma1_inv - sigma0_inv) / 2
    B = np.dot(sigma0_inv, mu0) - np.dot(sigma1_inv, mu1)
    c = (np.dot(np.dot(np.transpose(mu1), sigma1_inv), mu1) - np.dot(np.dot(np.transpose(mu0), sigma0_inv), mu0))/2
    c += log((1-pi)*np.linalg.det(sigma1) / (pi*np.linalg.det(sigma0)))
    
    return A, B, c

# Compute the decision function on one point

def f_decision_QDA(x, A_, B_, c_):
    # A label one should then be return if f>=0, else 0
    quadratic_term = np.dot(np.dot(np.transpose(x), A_), x)
    linear_term = np.dot(np.transpose(x), B_)
    cte_term = c_
    return quadratic_term + linear_term + cte_term

pi, mu0, mu1, sigma0, sigma1 = mle_QDA(features, labels)

A, B, c = learn_params_QDA(features, labels)
```
